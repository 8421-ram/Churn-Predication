```python
def UNA_outliers_remove(churn, var_group, include_outliers=False):
    size = len(var_group)
    plt.figure(figsize=(7*size, 4), dpi=100)
    
    for i, var in enumerate(var_group):
        quant25, quant75 = churn[var].quantile([0.25, 0.75])
        iqr = quant75 - quant25
        med = churn[var].median()
        whis_low = med - (1.5 * iqr)
        whis_high = med + (1.5 * iqr)

        if include_outliers:
            outliers_high = (churn[var] > whis_high) | (churn[var] < whis_low)
            filtered_churn = churn[~outliers_high]
        else:
            filtered_churn = churn
        
        plt.subplot(1, size, i+1)
        sns.boxplot(filtered_churn[var])
        plt.ylabel(var)
        plt.xlabel("{}".format(var), fontsize=20)
        plt.title("Without outliers" if not include_outliers else "With outliers")
```

```python
# columns divided  into the groups  
cust_details=["customer_id","vintage","age"]
current_month=["current_balance","current_month_credit","current_month_debit","current_month_balance"]
previous_month=["previous_month_balance","previous_month_credit","previous_month_end_balance",]
previous_quarters=["average_monthly_balance_prevQ","average_monthly_balance_prevQ2"]
trans_date=["day_of_last_transaction","month_of_last_tranaction","week_of_year_last_Transaction","dayof_week_last_tranaction"]
```

```python
def UNA_numeric(churn, variable_group):
    size=len(variable_group)
    plt.figure(figsize=(5*size,3),dpi=100)
    
    for j,i in enumerate(variable_group):
        mini=churn[i].min()
        maxi=churn[i].max()
        ran=churn[i].max()-churn[i].max()
        mean=churn[i].mean()
        median=churn[i].median()
        st_dev=churn[i].std()
        skew=churn[i].skew()
        kurt=churn[i].kurtosis()
        
        #varience
        points=mean-st_dev,mean+st_dev
        
        plt.subplot(1,size,j+1)
        sns.kdeplot(churn[i], shade=True)
       
        sns.lineplot(x=points, y=[0, 0], color="black", label="std_dev")

        sns.scatterplot(x=[mini,maxi], y=[0,0], color="pink", label="min_value,maxi_value")
        sns.scatterplot(x=[mean],y=[0],color="yellow",label="mean")
        sns.scatterplot(x=[median],y=[0],color="blue",label="median")
        plt.xlabel('xlabel', rotation=5)
        plt.xlabel("{}".format(i),fontsize=20)
        plt.ylabel("Frequency & count")
        plt.title("std_dev={}; kurtosis={};\nskew={};range={}\nmean={}; median={}".format((round(points[0],2),round(points[1],2)),
                                                                                          round(kurt,2),
                                                                                          round(skew),
                                                                                          (round(mini,2),round(maxi,2),round(ran,2)),
                                                                                          round(mean,2),
                                                                                          round(median,2)))
```

```python
def UNA_categorical(churn, variable_group):
    size = len(variable_group)
    plt.figure(figsize=(6*size, 3), dpi=100)
    
    for j, i in enumerate(variable_group):
        norm_count = churn[i].value_counts(normalize=True)*100
        nun_unique = churn[i].nunique()
        
        plt.subplot(1, size, j+1)
        sns.barplot(x=norm_count.index, y=norm_count, order=norm_count.index)
        plt.ylabel("percentage", fontsize=20)
        plt.xticks(rotation=90)
        plt.xlabel("{}".format(i), fontsize=20)
        plt.title('nunique={}\n value counts \n {}'.format(nun_unique, norm_count))
```
